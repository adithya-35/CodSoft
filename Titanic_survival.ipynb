```python
mean = df['Age'].mean()
df['Age'] = df['Age'].fillna(df['Age'].mean())
```

```python
df.drop(columns="Cabin", inplace=True)
```

```python
df.head()
```

```python
Survived=df['Survived'].value_counts().reset_index()
print(Survived)
```

```python
import seaborn as sns

sns.barplot(x='Survived', y='count', data=Survived)
plt.xlabel('Survived')
plt.ylabel('Count')
plt.title('Survived Counts')
plt.show()
```

```python
survived_gender = df.groupby(['Sex', 'Survived']).size().unstack()
```

```python
survived_gender.plot(kind='bar', stacked=True)
plt.xlabel('Sex')
plt.ylabel('Count')
plt.title('Survived Counts by Gender')
plt.xticks(rotation=0)
plt.legend(['Not Survived', 'Survived'])
plt.show()
```

```python
#Make a column of Age group give easy and good insights
# Define age ranges
age_bins = [0, 30, 60, float('inf')]  # Define age ranges: 0-30 (Younger), 31-60 (Middle), 61 and above (Older)
age_labels = ['Younger', 'Middle', 'Older']
```

```python
# Create age groups
df['Age_Group'] = pd.cut(df['Age'], bins=age_bins, labels=age_labels, right=False)
```

```python
df
```

```python
sns.countplot(data=df, x='Age_Group', order=['Younger', 'Middle', 'Older'])
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.title('Count of Individuals in Each Age Group')
plt.show()
```

```python
survived_counts = df.groupby(['Age_Group', 'Survived']).size().unstack()

survived_counts.plot(kind='bar', stacked=True)
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.title('Survival Status by Age Group')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()
```

```python
from inspect import FullArgSpec
#Make a column of Fare group give easy and good insights
# Define Fare ranges
Fare_bins = [0, 50, 100, float('inf')]  # Define Fare ranges: 0-50 (Low), 51-100 (Medium), 100 and above (high)
Fare_labels = ['Low', 'Medium', 'High']

# Create Fare groups
df['Fare_Group'] = pd.cut(df['Fare'], bins=Fare_bins, labels=Fare_labels, right=False)
```
