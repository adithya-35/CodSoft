Data Visualization

Sales by TV

```python
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='TV', y='Sales')
plt.title('Sales by TV')
plt.xlabel('TV')
plt.ylabel('Sales')
plt.show()
```

Sales Distribution

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Sales'], bins=20, kde=True)
plt.title('Distribution of Sales')
plt.xlabel('Sales')
plt.ylabel('Frequency')
plt.show()
```

Sales by Radio

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Radio', y='Sales')
plt.title('Sales by Radio')
plt.xlabel('Radio')
plt.ylabel('Sales')
plt.show()
```

Sales by Newspaper

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Newspaper', y='Sales')
plt.title('Sales by Newspaper')
plt.xlabel('Newspaper')
plt.ylabel('Sales')
plt.show()
```

Correlation Matrix

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
